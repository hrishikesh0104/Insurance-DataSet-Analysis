### Categorical vs Target Column

```python
# Categorical vs Target Column ( Sex vs Charges)
sns.boxplot(x = 'sex' , y = 'charges', data = df)
plt.show()
# It can be inferred that charges against female candidates is lesser than that of their male counterparts
```

```python
# Categorical vs Target Column ( Smoker vs Charges)
sns.boxplot(x = 'smoker' , y = 'charges', data = df)
plt.show()
# It can be understood that charges against smokers is higher than that of non-smokers
```

```python
# Categorical vs Target Column ( region vs Charges)
sns.boxplot(x = 'region' , y = 'charges', data = df)
plt.show()
# Overall charges in southeast region is the highest and lowest in the southwest region
```

### Numerical vs Target Column

```python
# Numerical vs Target Column (age vs Charges)
sns.scatterplot(x = 'age' , y = 'charges', data = df)
plt.show()
# It can be inferred that with age the charges increases
```

```python
# Numerical vs Target Column (bmi vs Charges)
sns.scatterplot(x = 'bmi' , y = 'charges', data = df)
plt.show()
# There seems to be very low relationship between bmi and charges
```

```python
# Numerical vs Target Column ( Smoker vs Charges)
sns.scatterplot(x = 'children' , y = 'charges', data = df)
plt.show()
# As the number of children increases the charges decrease
```
